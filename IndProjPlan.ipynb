```r
# Reading in the data frame
# Getting summary statistics (mean)
players <- read_csv("players.csv")
players
summarize(players,
          mean_played_hours = mean(played_hours),
          mean_Age = mean(Age, na.rm = TRUE))
```

I think that the data in the players.csv data frame is tidy already, and I do not think it needs minimal tidying or wrangling. It fits the criteria of having each row as being a single observation, each column being a single variable, and each cell holding only one value.
```r
# Exploratoring data with visualizations
# Exploring Age and subscription in players.csv

options(repr.plot.width=11, repr.plot.height=8)
playersAge_plot <- ggplot(players, aes(x= Age, fill = subscribe)) +
                    geom_bar(position = "stack") +
                    labs(x= "Player Age",
                         y= "Amount of Players Subscribed",
                         title = "Player age vs Proportion of Players Subscribed",
                          fill = "Subscribed") +
                    theme(text = element_text(size = 20))
playersAge_plot
```

From the graph above it seems that as players get older they are more likely to not be subscribed. It also appears there are much more players of younger ages that are subscribed. Particularily from ~16-25.

```r
# Exploring age of player, hours played and subscription status
options(repr.plot.width=11, repr.plot.height=8)
playtime_plot <- ggplot(players, aes(x= Age, y = played_hours, colour = subscribe)) +
                    geom_point() +
                    labs(x= "Player Age",
                         y= "Time spent playing (hours)",
                         colour = "Subscribed",
                         title = "Player age vs Proportion of Players Subscribed") +
                    theme(text = element_text(size = 20))
playtime_plot
```

With this graph, I attempted to visualize the way that player age could affect time spent playing on the Minecraft server. While there are people that playe 0 hours across all age ranges, it appears that the players who spend the most amount of hours on the game (hours ranging from 25-225) tend to be younger (ranging from 9-25). This graph also helps illustrate that the subscription status of the players based on their ages and time spent playing. It can be seen that younger players who are logging the most hours, described above, are more the ones who are subscribed.
