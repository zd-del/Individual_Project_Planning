### players.csv (players) description/summary statistics
- 196 observations (rows).
- 7 variables.
    - experience
      - Type of variable: character (chr)
      - This variable contains 5 categories and details experience level of each unique MineCraft player (Beginner, Amateur, Regular, Veteran, Pro).
    - subscribe 
      - Type of variable: logical (lgl)
      - This variable contains two categories and details if each unique player is subscribed to MineCraft (TRUE or FALSE). 
    - hashedEmail  
      - Type of variable: character (chr)
      - This variable details each unique players hashed email address. 
    - played_hours 
      - Type of variable: double (dbl)
      - This variable details the hours played by each individual on the MineCraft server.
    - name
      - Type of variable: character (chr)
      - This variable reports each player first name.
    - gender 
      - Type of variable: character (chr)
      - This reports the identified gender of each player.
    - Age 
      - Type of variable: double (dbl)
      - This variable reports the age of each player.
